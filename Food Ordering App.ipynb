```python
class FoodItem:
    def __init__(self, name, quantity, price, stock):
        self.food_id = None  
        self.name = name
        self.quantity = quantity
        self.price = price
        self.stock = stock

class RestaurantMenu:
        
    Menu = []

    def generate_food_id(self):
        import random
        import string
        random_numbers = ''.join(random.choices(string.digits, k=4))
        food_id = 'FI-' + random_numbers
        return food_id
    
    def add_food_item(self, name, quantity, price, stock):
        food_item = FoodItem(name, quantity, price, stock)
        food_item.food_id = self.generate_food_id()
        food_items = {}
        food_items[food_item.food_id] = food_item
        RestaurantMenu.Menu.append(food_items)

    def edit_food_item(self, food_id, new_name, new_quantity, new_price, new_stock):
        for food_items in RestaurantMenu.Menu:
            for food_id,food_item in food_items.items():
                if food_item.food_id == food_id:
                    food_item.name = new_name
                    food_item.quantity = new_quantity
                    food_item.price = new_price
                    food_item.stock = new_stock
                    break

    def remove_food_item(self, food_id):
        for food_items in RestaurantMenu.Menu:
            for food_id,food_item in food_items.items():
                if food_item.food_id != food_id:
                    RestaurantMenu.food_items[food_id] = food_item

    def view_all_food_items(self):
        return RestaurantMenu.Menu
        
class Admin(RestaurantMenu):
    def __init__(self, admin = None):
        super().__init__()
        self.adminset = {"adrij@49", "jyoti@77", "rishab@56"}
        self.admin = admin

    def get_admin(self,admin):
        if admin in self.adminset:
            print("Welcome!!!")
        else:
            print("No such Admin is there  :( ")
    
    def get_adminset(self):
        return self.adminset

    def add_admin(self, admin):
        self.adminset.add(admin)

    def remove_admin(self, admin):
        self.adminset.remove(admin)

    def modify_admin(self, old_admin, new_admin):
        if old_admin in self.adminset:
            self.adminset.remove(old_admin)
            self.adminset.add(new_admin)
        else:
            print(f"{old_admin} does not exist in the adminset.")
            
class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
    
    def __eq__(self, other):
        if isinstance(other, User):
            return self.full_name == other.full_name and self.phone_number == other.phone_number and self.email == other.email and self.password == other.password
        return False
        
class Application:
    def __init__(self):
        self.restaurant_menu = RestaurantMenu()
        self.users = []

    def register(self, full_name, phone_number, email, address, password):
        user = User(full_name, phone_number, email, address, password)
        self.users.append(user)

    def login(self, full_name, password):
        for user in self.users:
            if user.full_name == full_name and user.password == password:
                self.run()
            else:
                print("Incorrect User ID and Password. Please try again or Register.")

    def update_profile(self):
        full_name = input("Current Name: ")
        phone_number = input("Current Phone Number: ")
        email = input("Current Email-ID: ")
        address = input("Current Address: ")
        password = input("Current Password: ")
        user = User(full_name, phone_number, email, address, password)
        print("\n")
        for u in self.users:
            if u == user:
                u.full_name = input("New Name: ")
                u.phone_number = input("New Phone Number: ")
                u.email = input("New Email-ID: ")
                u.address = input("New Address: ")
                u.password = input("New Password: ")
                print("\n")
            else:
                print("\n")
                print("ERROR: invalid Input. Try Again!")
                print("\n")
                
    def run(self):
        print("-"* 20 + "Please Select Option" + "-"* 20)
        print(" "* 15 + "To place a New Order Enter 'O'")
        print(" "* 14 + "To check Order History Enter 'H'")
        print(" "* 16 + "To Update Profile Enter 'U'")
        print(" "* 17 + "To Exit the App Enter 'E'")
        print("\n")
        input_opp = str(input("Please Select Your Operation: ")).upper()
        if (input_opp == 'O'):
            print("\n" * 2)
            new_order.place_order()
            print("\n")
            self.run()
        elif (input_opp == 'H'):
            new_order.print_order_history()
            self.run()
        elif (input_opp == 'U'):
            print("\n" * 2)
            self.update_profile()
            self.run()
        elif (input_opp == 'E'):
            print("\n")
            print("-" * 15 + "Thank You For Using This App" + "-" * 15)
            return
        else:
            print("\n" * 2 + "ERROR: Invalid Input. Try again!")
            input_opp = str((input("Please Select Your Operation: "))).upper()
            
class Place_New_Order():
        
    ordered_items = []
    order_history = []
    
    def display_menu(self):
        print("-" * 25 + "Menu:" + "-" * 25)
        menu = RestaurantMenu()
        for food_items in menu.Menu:
            for food_id, food_item in food_items.items():
                print(f"{food_item.food_id}    {food_item.name}   {food_item.quantity}   [INR {food_item.price}]")

    def find_item(self, foodid):
        menu = RestaurantMenu()
        for food_items in menu.Menu:
            for food_id, food_item in food_items.items():
                if food_item.food_id == foodid:
                    return food_item.name 
        return None
                
    def place_order(self):
        self.display_menu()
        print("\n" * 2)
        menu = RestaurantMenu()
        while True:
        
            food_id0 = str(input("Enter the Food ID of the item you want to order (or 'Q' to quit): ")).upper()

            if food_id0 == 'Q':
                break
            
            quantity = int(input("Enter the quantity: "))
            
            order_item = {}
            ordered_item = self.find_item(food_id0)
            if ordered_item is None:
                print("Item not found in the menu. Please try again.")
                continue
            
            for food_items in menu.Menu:
                for food_id, food_item in food_items.items():
                    if food_item.stock >= quantity:
                        order_item['Name'] = ordered_item
                        order_item['Quantity'] = quantity
                        food_item.stock = food_item.stock - quantity
                    else:
                        print("Out of Stock.\n")
                        self.place_order()
            
            order_item['Price'] = self.return_price(food_id0)
            Place_New_Order.ordered_items.append(order_item)
            print(f"{ordered_item} ({quantity}) added to your order.\n")            
        print("\nOrder Summary:\n")
        self.print_order_summary()

        Place_New_Order.order_history.append(Place_New_Order.ordered_items.copy())
        Place_New_Order.ordered_items.clear()
        print("\nYour order has been placed successfully!")
    
    def return_price(self, foodid):
        menu = RestaurantMenu()
        for food_items in menu.Menu:
            for food_id, food_item in food_items.items():
                if food_item.food_id == foodid:
                    return food_item.price
    
    def print_order_summary(self):
        total_cost = 0
        for item in Place_New_Order.ordered_items:
            item_name = item['Name']
            item_price = item['Price']
            quantity = item['Quantity']
            item_total = item_price * (quantity)
            total_cost += item_total
            print(f"{item_name} x {quantity} = ₹{item_total}")
        
        if total_cost >= 1000:
            total_cost = total_cost - (total_cost * 0.05)
            print("Discount Criteria Met!!! Discount of 5% Applied")
            print(f"\nTotal: ₹{total_cost}")
        elif total_cost >= 3000:
            total_cost = total_cost - (total_cost * 0.1)
            print("Discount Criteria Met!!! Discount of 10% Applied")
            print(f"\nTotal: ₹{total_cost}")
        elif total_cost >= 5500:
            total_cost = total_cost - (total_cost * 0.15)
            print("Discount Criteria Met!!! Discount of 15% Applied")
            print(f"\nTotal: ₹{total_cost}")
        else:
            total_cost = total_cost
            print(f"\nTotal: ₹{total_cost}")

    def print_order_history(self):
        print("\n" * 2)
        print("-" * 25 + "Order History:" + "-" * 25)
        history_string = ""
        for order_num in Place_New_Order.order_history:
            for order_item in order_num:
                history_string += f"Name: {order_item['Name']}   Quantity: {order_item['Quantity']}   Price: [ INR {order_item['Price']} ]\n"
            history_string += "\n"
            
        print(history_string)
        print("\n")
        

new_order = Place_New_Order()
```
